```python
df = pd.read_csv("http://51.91.120.89/extras/nba-ok.csv")
df
```

```python
from scipy.stats import trim_mean
```

```python
def tm(x):
    return trim_mean(x, 0.1)
```

```python
dfg = df.groupby("Team")
```

```python
dfg["Salary"].agg(["mean", tm])
```

```python
dfg.agg({
    "Age" : [min, max],
    "Salary": ["mean", tm, "median"]
})
```
