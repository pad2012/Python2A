```python
html_data = bs4.BeautifulSoup(response.text)
```

```python
tbody = html_data.find("tbody", class_="lister-list")
```

```python
rows = tbody.find_all("tr")
```

```python
result = []
for row in rows:
    ranking = row.find("span").get("data-value") 
    print("ranking:", ranking)
    
    title =  row.find("td", class_="titleColumn").find("a").text
    print("tytuł:", title)
    
    link = row.find("td", class_="titleColumn").find("a").get("href")
    print("link:", link )
    
    year = row.find("td", class_="titleColumn").find("span").text[1:-1] 
    print("rok:", year)
    
    score = row.find("td", class_="imdbRating").text.strip()
    print("ocena:", score )
    
    users = row.find("td", class_="imdbRating").find("strong").get("title").split(" ")[-3].replace(",","") 
    print("liczba ocen:", users)
    result.append({
        "ranking": ranking, "title": title, "year":year, "score":score, "users":users, "link":link
    })
    print("="*50)
```
