```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 4.5*np.pi, 0.2)
x
```

```python
y = np.sin(x)
y
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y, c="red", linestyle="--", marker="d")
plt.grid(axis="y")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Wykres sin(x)")
plt.xticks( np.arange(0,4.5*np.pi,np.pi), ['0','π','2π','3π','4π'] )
plt.show()
```

```python
# wykres punktowy
# (1,1) (2,5) (3,8) (4,10)
plt.plot([1,2,3,4], [1,5,8,10], "r^")
plt.axis([0,5,0,15])
plt.show()
```

```python
x = np.arange(0, 5.2, 0.2)
plt.figure(figsize=(10,6))
plt.plot(x, x, "g--")
plt.plot(x, 1/2*x**2, "k-.")
plt.plot(x, 1/8*x**3, "r")
plt.text(0,12, "Teskt na wykresie", fontsize=30, color="gray")
plt.annotate("Adnotacja na wykresie", [2,2], [3,10], arrowprops={
    "linewidth":1, "arrowstyle":"<->"
})
plt.show()
```

```python
plt.plot(x, x, "g--",  x, 1/2*x**2,"k-.",   x, 1/8*x**3,"r"  )
```

### Wykres słupkowy

```python
countries = ["USA","Rosja","Kanada"]
m_bronze = np.array([12, 10, 8])
m_silver = np.array([4,2,8])
m_gold = np.array([2,2,1])



plt.bar(countries, m_bronze, width=0.5, color="brown", label="B")
plt.bar(countries, m_silver, width=0.5, color="silver", bottom=m_bronze, label="S")
plt.bar(countries, m_gold, width=0.5, color="gold", bottom=m_bronze+m_silver, label="G")
plt.legend(loc="upper right", ncol=2)
plt.title("Statystyka medalowa")

#plt.figure(figsize=(10,6))
plt.gcf().set_size_inches(10,6)

# zapis do pliku
plt.savefig("wykres.jpg", dpi=150)
plt.savefig("wykres.png", dpi=75)

plt.show()
```

```python
x = np.array([1,5,9])
countries = ["USA","Rosja","Kanada"]
m_bronze = np.array([12, 10, 8])
m_silver = np.array([4,2,8])
m_gold = np.array([2,2,1])


plt.bar(x, m_bronze, width=0.5, color="brown", label="B")
plt.bar(x+0.5, m_silver, width=0.5, color="silver",  label="S")
plt.bar(x+1, m_gold, width=0.5, color="gold",  label="G")
plt.legend(loc="upper right", ncol=2)
plt.title("Statystyka medalowa")

#plt.figure(figsize=(10,6))
plt.gcf().set_size_inches(10,6)

plt.xticks(x+0.5, countries)

plt.show()
```

### Wykres kołowy

```python
names = ["Jaś","Małgosia","Zuzia","Julka"]
toys = [10, 7, 18, 5]
_colors = ["blue", "yellowgreen", "#FF0033", "lightskyblue"]

plt.figure(figsize=(9,7))
plt.pie(toys, labels=names, colors=_colors, shadow=True, autopct="%.1f", explode=[0.1, 0, 0, 0.2])
plt.show()
```

### Wykres rozrzutu wartości (scatter plot)

```python
np.random.seed(0)
N = 100
x = 6 * np.random.randn(N) - 3
y = 2*x + np.random.randn(N)*5
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x, y, c="r", marker="s")
```
