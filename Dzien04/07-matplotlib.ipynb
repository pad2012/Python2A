```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 4.5*np.pi, 0.2)
x
```

```python
y = np.sin(x)
y
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y, c="red", linestyle="--", marker="d")
plt.grid(axis="y")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Wykres sin(x)")
plt.xticks( np.arange(0,4.5*np.pi,np.pi), ['0','π','2π','3π','4π'] )
plt.show()
```
