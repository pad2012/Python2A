```python
url = "https://stooq.pl/q/d/l/?s=cdr&d1=20200101&d2=20220114&i=d"
df = pd.read_csv(url, parse_dates=True, index_col="Data")
df.head(20)
```

```python
df.Zamkniecie.plot(figsize=(16,6), grid=True, color="red")
```

```python
df.Zamkniecie.loc["2021-10-01":"2022-01-14"].plot(figsize=(11,6), grid=True, color="red")
```

```python
df.Zamkniecie.loc["2021-12"].plot(figsize=(11,6), grid=True, color="red")
```

```python
df.Zamkniecie.loc["2021":].plot(figsize=(11,6), grid=True, color="red")
```

```python
# średnie notowania roczne
df.Zamkniecie.resample("AS").mean().plot(figsize=(11,6), grid=True, color="red", kind="bar")
```

```python
# maksymalnie notowania kwartalnie
df.Zamkniecie.resample("QS").max().plot(figsize=(11,6), grid=True, color="red", kind="bar")
```

```python
df.Zamkniecie.resample("14D").mean().plot(figsize=(6,15), grid=True, color="red", kind="barh")
```

```python
# obliczanie i wizualizacja średniej kroczącej
line1 = df.Zamkniecie
line1.name = "Dane oryg."
line1.plot(color="red", legend=True)

line2 = df.Zamkniecie.rolling(30).mean()
line2.name = "MAvg30"
line2.plot(figsize=(16,6), grid=True, color="green", legend=True)
```
