```python
import pandas as pd
```

```python
url = "https://stooq.pl/q/d/l/?s=pkn&d1=20211201&d2=20220114&i=d"
df =pd.read_csv(url)
df
```

```python
df["Zmiana_1D"] = pd.np.NaN
df.head()
```

```python
%%timeit -r 10 -n 10
for i in range(1,df.shape[0]):
    df.iloc[i, -1] = df.iloc[i, -3] - df.iloc[i-1, -3]
```

```python
df.head()
```

```python
%timeit -r 10 -n 10 df["Zmiana_1D_shift"] = df.Zamkniecie - df.Zamkniecie.shift(1)
```
