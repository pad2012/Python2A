```python
import numpy as np
import pandas as pd
```

### Dyskretyzacja

```python
np.random.seed(0)
age = np.random.randint(0, 106, 40)
age
```

```python
df = pd.DataFrame(age, columns=['Wiek'])
df
```

```python
# dyskretyzacja
df["Zakres wiekowy"] = pd.cut(df.Wiek, bins=[0, 18, 65, 110], right=False, labels=["Dzieci","Dorośli","Seniorzy"])
df.head(10)
```
