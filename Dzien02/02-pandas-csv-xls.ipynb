```python
# wymuszanie typu dla importowanych kolumn
# ładowanie pliku CSV do obiektu DataFrame
url = "http://51.91.120.89/extras/f500.csv"
df = pd.read_csv(url, dtype={
    "rank" : "uint16", 
    "revenue_change" : "float32"
})
df.dtypes
```

```python
# konwertowanie importowanych kolumn
# ładowanie pliku CSV do obiektu DataFrame
url = "http://51.91.120.89/extras/f500.csv"
df = pd.read_csv(url, converters={
    "rank" : lambda x: int(x)<=10
} )
df.head(15)
```
