```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("http://51.91.120.89/extras/nba.csv", index_col='Name')
df
```

```python
# szukamy NaN
df.isna().sum()
```

```python
# usuwamy wiersze zawierajce wszędzie wartości NaN
df.dropna(axis=0, how='all', inplace=True)
df
```

```python
df.isna().sum()
```

```python
# zamiana NaN w College na string pusty
df["College"].replace(np.NaN, "", inplace=True)
```

```python
df
```

```python
df.dtypes
```

```python
# konwersja funty -> kg
df.Weight /= 2.2
```

```python
def convert_height(s):
    feet, inches = s.split("-")
    return int(feet)*30.4 + int(inches)*2.54
```

```python
convert_height("7-1")
```

```python
df.Height = df.Height.apply(convert_height)
```

```python
# zapisz do CSV
df.to_csv("nba1.csv")
```
